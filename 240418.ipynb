```python
model = Sequential()
model.add(Dense(train_data.shape[1], input_dim = train_data.shape[1]))
model.add(LSTM(128, activation='tanh'))
model.add(Dropout(0.25))
model.add(LSTM(32, activation='tanh'))
model.add(Dropout(0.25))
model.add(Dense(1))

model.summary()
```
